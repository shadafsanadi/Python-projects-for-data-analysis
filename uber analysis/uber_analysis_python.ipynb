Time Series Analysis
First, we'll convert the tpep_pickup_datetime and tpep_dropoff_datetime columns to datetime objects and extract relevant time features.

```python
df['tpep_pickup_datetime']=pd.to_datetime(df['tpep_pickup_datetime'])
df['tpep_dropoff_datetime'] = pd.to_datetime(df['tpep_dropoff_datetime'])
```

```python
df['pickup_date'] = df['tpep_pickup_datetime'].dt.date
df['pickup_hour'] = df['tpep_pickup_datetime'].dt.hour
df['pickup_dayof_week'] = df['tpep_pickup_datetime'].dt.dayofweek
df['dropoff_date'] = df['tpep_dropoff_datetime'].dt.date
df['dropoff_hour'] = df['tpep_dropoff_datetime'].dt.hour
df['dropoff_dayof_week'] = df['tpep_dropoff_datetime'].dt.dayofweek

```

```python
import seaborn as sns
```

```python
df['trip_duration'] = (df['tpep_dropoff_datetime'] - df['tpep_pickup_datetime']).dt.total_seconds() / 60
df['trip_duration']
```

```python
df['trip_duration'] = (df['tpep_dropoff_datetime'] - df['tpep_pickup_datetime']).dt.total_seconds() / 60

# Plot the distribution of trip duration using Seaborn
plt.figure(figsize=(12, 6))
sns.histplot(df['trip_duration'], bins=50, kde=False, binrange=(0, 100))
plt.title('Distribution of Trip Duration')
plt.xlabel('Duration (minutes)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.hist(df['fare_amount'], bins=50,range=(0, 100))
plt.title('Distribution of Fare Amount')
plt.xlabel('Fare Amount ($)')
plt.ylabel('Frequency')
plt.show()
```

```python
total_revenue = df['total_amount'].sum()
print(f'Total Revenue: ${total_revenue:.2f}')
```

```python
average_revenue_per_trip = df['total_amount'].mean()
print(f'Average Revenue per Trip: ${average_revenue_per_trip:.2f}')
```

```python
rides_per_passenger = df['passenger_count'].value_counts()
rides_per_passenger = rides_per_passenger.reset_index()
rides_per_passenger
```

```python
rides_per_passenger.plot(kind='bar', figsize=(12, 6), title='Number of Rides per Passenger Count')
plt.xlabel('Passenger Count')
plt.ylabel('Number of Rides')
plt.show()
```

```python
top_pickup_locations = df.groupby(['pickup_latitude', 'pickup_longitude']).size().nlargest(10)
print('Top 10 Pickup Locations:')
print(top_pickup_locations)

# Top 10 dropoff locations
top_dropoff_locations = df.groupby(['dropoff_latitude', 'dropoff_longitude']).size().nlargest(10)
print('Top 10 Dropoff Locations:')
print(top_dropoff_locations)
```

```python
pickup_heatmap_data = df.pivot_table(
    index=df['tpep_pickup_datetime'].dt.dayofweek,
    columns=df['tpep_pickup_datetime'].dt.hour,
    values='VendorID',
    aggfunc='count'
)

# Plot heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(pickup_heatmap_data, cmap="YlGnBu")
plt.title('Heatmap of Pickups by Hour and Day of Week')
plt.xlabel('Hour of Day')
plt.ylabel('Day of Week')
plt.show()
```

```python
avg_fare_by_payment_type = df.groupby('payment_type')['fare_amount'].mean().reset_index()

# Plot bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x='payment_type', y='fare_amount', data=avg_fare_by_payment_type)
plt.title('Average Fare Amount by Payment Type')
plt.xlabel('Payment Type')
plt.ylabel('Average Fare Amount ($)')
plt.show()
```

```python
#hourley revenue 

df['pickup_hour'] = df['tpep_pickup_datetime'].dt.hour
hourley_revenue = df.groupby('pickup_hour')['total_amount'].mean().reset_index()
hourley_revenue
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='pickup_hour', y='total_amount', data=hourley_revenue, palette='viridis')
plt.title('Average Revenue by Pickup Hour')
plt.xlabel('Pickup Hour')
plt.ylabel('Average Total Amount ($)')
plt.xticks(range(24))  # Ensure all hours from 0 to 23 are shown on the x-axis
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='trip_distance', y='trip_duration', data=df, alpha=0.3)
plt.title('Trip Distance vs. Trip Duration')
plt.xlabel('Trip Distance (miles)')
plt.ylabel('Trip Duration (minutes)')
plt.show()
```

```python
#Trips by Day of the Week
df['day_of_week'] = df['tpep_pickup_datetime'].dt.dayofweek

# Calculate the number of trips per day of the week
trips_by_day_of_week = df['day_of_week'].value_counts().sort_index()

# Plot the number of trips per day of the week
plt.figure(figsize=(12, 6))
sns.barplot(x=trips_by_day_of_week.index, y=trips_by_day_of_week.values)
plt.title('Number of Trips by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Trips')
plt.xticks(ticks=range(7), labels=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.show()
```

```python
# trip speed analysis
df['trip_duration_hours'] = df['trip_duration'] / 60
df['trip_speed'] = df['trip_distance'] / df['trip_duration_hours']

# Plot the distribution of trip speeds
plt.figure(figsize=(12, 6))
sns.histplot(df['trip_speed'].dropna(), bins=500, kde=True)
plt.title('Distribution of Trip Speeds')
plt.xlabel('Speed (miles per hour)')
plt.ylabel('Frequency')
plt.show()


```
